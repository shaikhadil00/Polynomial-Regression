```python
x = df.iloc[:,1:-1].values
y = df.iloc[:,-1].values
```

# The above code is all abt selecting specific column 

# In this we select experience vs income

```python
x
```

```python
y
```

```python
from sklearn.preprocessing import PolynomialFeatures
poly_reg = PolynomialFeatures(degree=2)
x_poly = poly_reg.fit_transform(x)
```

```python
x_poly
```

 There will be Two Extra column (1column = "1" , 2column = Exprience , 3column = squared of Exprience column)

# Because of Degree = 2 , Therefore we are getting 2 Extra column .

# As per degree increase column also increase

```python
from sklearn.linear_model import LinearRegression
reg = LinearRegression()
reg.fit(x_poly,y)
```

```python
plt.scatter(x,y)
plt.plot(x,reg.predict(poly_reg.fit_transform(x)))
plt.show()
```

```python
reg.predict(poly_reg.fit_transform([[11]]))
```

```python
reg.score(x_poly,y)
```
